Video to images

```python
import cv2
import re
from pathlib import Path
from tqdm import tqdm
```

```python
video = cv2.VideoCapture('./test.webm')
if not video.isOpened():
    print("Video not file not found / Error in loading video")

folder = Path('./Images/')
if not folder.exists():
    folder.mkdir(parents=True, exist_ok=True)
else:
    for file in folder.iterdir():
        if file.is_file():
            file.unlink()

frame_no = 0
while True:
    ret,frame = video.read()
    if not ret:
        break
    file_name = './Images/' + str(frame_no)+ '.jpg'
    cv2.imwrite(file_name, frame)
    frame_no += 1


video.release()
cv2.destroyAllWindows()
```

Images to Video

```python
def extract_number(filename):
    numbers = re.findall(r'\d+', filename.stem)
    return int(numbers[0]) if numbers else None

input_images_path = Path('./Images/')
output_video_path = Path('./output_video.webm')

images = list(input_images_path.glob('*.jpg'))  # Assuming the images are in JPG format. Change accordingly.
images.sort(key=extract_number)

frame = cv2.imread(str(images[0]))
height, width, layers = frame.shape

frame_rate = int(input("Enter frame rate: "))

video = cv2.VideoWriter(str(output_video_path), cv2.VideoWriter_fourcc(*'VP90'), frame_rate, (width, height))

for image_path in tqdm(images):
    print(f"Processing {image_path.name}")
    read_image = cv2.imread(str(image_path))
    read_image = cv2.resize(read_image, (width, height))
    video.write(read_image)

video.release()
cv2.destroyAllWindows()
```
